```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def enviar_email(destinatario, commits):
    # Configurações do servidor SMTP do Gmail
    smtp_server = "smtp.gmail.com"
    smtp_port = 587
    email_remetente = ""
    senha_remetente = ""

    # Criação do objeto de mensagem
    msg = MIMEMultipart()
    msg['From'] = email_remetente
    msg['To'] = destinatario
    msg['Subject'] = "Code refactoring - Master's Research"
    mensagem = f"""Hello developer!

    We are conducting research that aims to create an index of refactoring triviality in software projects. Our goal is to predict how trivial it is to perform a refactoring operation.

    Some benefits are: assisting developers in allocating time, facilitating decisions about prioritizing resources, helping to assess risks associated with changes, evaluating changes in production modules, in addition to encouraging the use of refactoring as a good practice in software development. .

    We noticed that you frequently applied refactoring operations in some projects we analyzed, which indicates to us that you value the quality and maintenance of code. Therefore, we would like your contribution to the creation of this index by answering some questions:

    1. Did you intend to apply refactoring to the commits below?
    2. If yes, what operations did you want to perform on each commit?
    3. Could you answer a survey about the importance of metrics in refactoring activities? (https://forms.gle/HTStLFXyRJdtsZmMA)

    Commits:
    {commits}

    (If you agree to participate, as a reward, at the end of the study we can apply our index to one of your projects).

    Thank you in advance for your attention and I am available for any further clarification.
    I await your return.

    Darwin Pinheiro
    Federal University of Ceara
    Master's student in Computer Science
    """

    # Adiciona o corpo da mensagem
    msg.attach(MIMEText(mensagem, 'plain'))

    try:
        # Conectando ao servidor SMTP do Gmail
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()  # Inicia a conexão TLS
        server.login(email_remetente, senha_remetente)  # Faz login no servidor
        server.sendmail(email_remetente, destinatario, msg.as_string())  # Envia o e-mail
        server.close()  # Encerra a conexão
        print(f"SUCESSO - {destinatario} - E-mail enviado com sucesso!")
    except Exception as e:
        print(f"ERRO - {destinatario} - Erro ao enviar e-mail: {e}")


# commit= 'XX'
# projeto= 'YYY'
# # Exemplo de uso
# destinatario = "darwinop@gmail.com"

# enviar_email(destinatario, assunto, mensagem)
```

```python
#importar json
import json
import time
from random import randint

#Abrir arquivo json
dataset='apache'
with open(f'./2_output_apiGithub/output-dev-{dataset}.json') as f:
    data = json.load(f)

#executar função a cada intervalo de tempo randomico de 10~30segundos
for obj in data:
    enviar_email(obj['developer_name'], obj['commits'])
    time.sleep(randint(10, 20)) 
```
